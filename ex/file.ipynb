```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer
```

```python
# Define the CSV file path
csv_file = "Data.csv"

# Read the CSV file into a pandas DataFrame
dataset = pd.read_csv(csv_file)

# Print the DataFrame
dataset
```

```python
new_dataset = dataset.drop(["Unnamed: 0","Unnamed: 5"], axis='columns')
new_dataset
```

```python
df = new_dataset.rename(columns={'Unnamed: 1': 'Country', 'Unnamed: 2': 'Age', 'Unnamed: 3': 'Salary', ' ': 'Purchased'})
df = df.drop(0)
df
```

```python
df_dict = {
    'yes' : 1,
    'No' : 0
}

df = df.replace(df_dict.keys(),df_dict.values())

df['c1'], df['c2'], df['c3']= 0, 0, 0
df.loc[df["Country"] == "India", "c3"] = 1
df.loc[df["Country"] == "Sri lanka", "c2"] = 1
df.loc[df["Country"] == "China", "c1"] = 1




```

```python
df = df.drop(["Country"], axis='columns')
```

```python
imputer = KNNImputer(n_neighbors=5 , weights='distance')
df = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)
df
```

```python
df['Age'] = df['Age'].subtract(21).div(55-21)
df['Salary'] = df['Salary'].subtract(20000).div(98000-20000)
df = df[['c1', 'c2', 'c3', 'Age', 'Salary', 'Purchased']]
df
```
