```python
import numpy as np
from scipy.cluster import hierarchy
from matplotlib import pyplot as plt
from scipy.spatial import distance_matrix
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs

%matplotlib inline
```

```python
#Generate data

Data, cl = make_blobs(n_samples=15, centers=[[4,4], [-2, -1], [1, 1], [10,4]], cluster_std=0.9)

#print(f"Data : \n {Data}")
```

```python
plt.scatter(Data[:, 0], Data[:, 1], marker='o',c=cl, s=2, cmap='viridis')
```

```python
agglom = AgglomerativeClustering(n_clusters= 4, linkage= "complete")
# singles means MIN method of hieracical clustring
# average means the average method of hieracical clustring
# complete means the MAX method of hieracical clustring
agglom.fit(Data)

newlables = agglom.labels_
print(f"cl : {cl}")
print(f"labels : {newlables}")
```

```python
newlabels=agglom.labels_
plt.subplot(2, 1, 1)
plt.scatter(Data[:, 0], Data[:, 1], c=cl, s=2, cmap='viridis')
plt.subplot(2, 1, 2)
plt.scatter(Data[:, 0], Data[:, 1], c=newlabels, s=2, cmap='viridis')
```

```python
dist_matrix = distance_matrix(Data,Data)
Z = hierarchy.linkage(dist_matrix,"complete")
dendro = hierarchy.dendrogram(Z)
```

```python
dist_matrix = distance_matrix(Data,Data)
plt.subplot(3, 3, 1)
plt.title("average")
V = hierarchy.linkage(dist_matrix,"average")
dendro_V = hierarchy.dendrogram(V)
plt.subplot(3, 3, 2)
plt.title("complete")
Z = hierarchy.linkage(dist_matrix,"complete",)
dendro = hierarchy.dendrogram(Z)
plt.subplot(3, 3, 3)
plt.title("Single")
S = hierarchy.linkage(dist_matrix,"single",)
dendro = hierarchy.dendrogram(S)
```

```python
from sklearn import datasets
n_samples =2000
np.random.seed(0)
Data, labels = datasets.make_moons(n_samples, noise=0.10)
plt.scatter(Data[:, 0], Data[:, 1], c=labels, s=2, cmap='viridis')
```

```python
agglom = AgglomerativeClustering(n_clusters= 4, linkage= "complete")
agglom.fit(Data)
newlables = agglom.labels_
```

```python
newlables = agglom.labels_
plt.subplot(2, 1, 1)
plt.scatter(Data[:, 0], Data[:, 1], c=labels, s=2, cmap='viridis')
plt.subplot(2, 1, 2)
plt.scatter(Data[:, 0], Data[:, 1], c=newlables, s=2, cmap='viridis')
```
